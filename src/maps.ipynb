```python
import numpy as np
import pysal as ps
import random as rdm
from pysal.contrib.viz import mapping as maps
```

```python
shp_link = ps.examples.get_path("south.shp")
values = np.array(ps.open(ps.examples.get_path('south.dbf')).by_col("HR60"))
```

```python
maps.plot_choropleth(shp_link, values, 'quantiles')
```

```python
maps.plot_choropleth(shp_link, values, 'quantiles', savein="quantiles.png")
```

```python
maps.plot_choropleth(shp_link, values, 'quantiles', savein="quantiles.png", dpi=600)
```

```python
maps.plot_choropleth(shp_link, values, 'fisher_jenks', savein="fj.png", dpi=600)
```
