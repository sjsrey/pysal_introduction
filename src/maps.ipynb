```python
import numpy as np
import pysal as ps
import random as rdm
from pysal.contrib.viz import mapping as maps
```

```python
shp_link = ps.examples.get_path("south.shp")

years = range(60,100,10)
for year in years:
    v = "HR%d"%year
    out_name = "quantiles_%s.png"%v
    values = np.array(ps.open(ps.examples.get_path('south.dbf')).by_col(v))
    maps.plot_choropleth(shp_link, values, 'quantiles', savein=out_name, dpi=600, title=v)
    
```
