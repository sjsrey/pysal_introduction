```python
import numpy as np
import pysal as ps
import random as rdm
from pysal.contrib.viz import mapping as maps
```

```python
shp_link = ps.examples.get_path("south.shp")
w = ps.queen_from_shapefile(shp_link)
```

```python
hr = []
for year in range(60,100,10):
    v = "HR%d"%year
    values = np.array(ps.open(ps.examples.get_path('south.dbf')).by_col(v))
    hr.append(values)
```

```python
hr = np.array(hr).T
```

```python
hr.shape
```

```python
fv = np.ones((w.n,1), float)
```

```python
res = ps.Maxp(w, hr, floor=50, floor_variable=fv)
```

```python
res.area2region.values()
```

```python
region_ids = np.array(res.area2region.values())
region_ids.shape = (w.n,)

maps.plot_choropleth(shp_link, region_ids, 'unique_values', savein='regions.png', dpi=600, title='Maxp')
```
